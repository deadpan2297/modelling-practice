Here I will show how to do PCA and LDA on the on the IRIS data set

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
#Load the data. We only need the features for PCA, not labels.
iris = datasets.load_iris()
X = iris.data
```

```python
#Create the PCA object and project the data onto the PC axes
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
#Plotting
plt.figure()
colors = ["indigo", "orchid", "lightpink"]
lw = 1
y = iris.target
for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(
        X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, lw=lw, label=target_name
    )

plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA of IRIS dataset")
```

```python
#Create the LDA object and apply to the data
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)
#Plotting
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(
        X_r2[y == i, 0], X_r2[y == i, 1], alpha=0.8, color=color, label=target_name
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")
plt.show()
```

```python
# Percentage of variance explained for each components
print(
    "explained variance ratio (first two components): %s"
    % str(pca.explained_variance_ratio_)
)
```
